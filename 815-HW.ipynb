To compare the tightness of linear relaxations of 2 forms of constraints.

We can take a small example to take a look at this problem.
Set $I = {1, 2}$. 

The first linear relaxation becomes 
\begin{align*}
    y_1 &\geq y_0 \\
    y_2 &\geq y_0 \\
    y_1 + y_2 -1 &\leq y_0 \\
    y_0, y_1, y_2 &\in [0,1]
\end{align*}

the second linear relaxation becomes
\begin{align*}
    y_1 + y_2 &\geq 2y_0 \\
    y_1 + y_2 -1 &\leq y_0 \\
    y_0, y_1, y_2 &\in [0,1]
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

y0 = 0.7
y1 = np.arange (0, 1.01, 0.01)
y2 = np.arange (0, 1.01, 0.01)
plt.plot (y1, 1 + y0 - y1, label ='$y_1+y_2 -1 \leq y_0$')
plt.vlines(0, 0, 1)
plt.hlines(0, 0, 1)
plt.vlines(0.7, 0, 1)
plt.hlines(0.7, 0, 1)
plt.vlines(1, 0, 1)
plt.hlines(1, 0, 1)
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
x = np.arange (0.7, 1, 0.01)
yy1 = 1 + y0 - x
plt.fill_between(x, yy1, 1, color='grey', alpha='0.5')
plt.legend()
plt.show()
```

The area of feasible space in the 1st form is

\begin{align*}
    s_1 &= \frac{1}{2}(1-y_0)^2 \\
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

y0 = 0.7
y1 = np.arange (0, 1.01, 0.01)
y2 = np.arange (0, 1.01, 0.01)
plt.plot (y1, 1 + y0 - y1, label ='$y_1+y_2 -1 \leq y_0$')
plt.plot (y1, 2 * y0 - y1, label ='$y_1+y_2 \geq 2y_0$')
plt.vlines(0, 0, 1)
plt.hlines(0, 0, 1)
plt.vlines(1, 0, 1)
plt.hlines(1, 0, 1)
plt.xlabel('$y_1$')
plt.ylabel('$y_2$')

x1 = np.arange (0.4, 0.71, 0.01)
yy1 = 2 * y0 - x1
plt.fill_between(x1, yy1, 1, color='grey', alpha='0.5')

x2 = np.arange (0.7, 1, 0.01)
yy2 = 1 + y0 - x2
yy3 = 2 * y0 - x2
plt.fill_between(x, yy3, yy2, color='grey', alpha='0.5')

plt.legend()
plt.show()
```

The area of feasible space in the 1st form is

\begin{align*}
    s_2 &= \frac{1}{2}(2y_0-2)^2 - \frac{1}{2}(1-y_0)^2 = \frac{3}{2}(1-y_0)^2, y_0\in[0.5,1] \\
    s_2 &= 1 - \frac{1}{2}(2y_0)^2 - \frac{1}{2}(1-y_0)^2 = 1 - 2y_0^2 - \frac{1}{2}(1-y_0)^2, y_0\in[0,0.5] \\
\end{align*}


```python
y0 = np.arange (0, 1.01, 0.01)
s1 = 1/2 * (1 - y0) ** 2

y01 = np.arange (0, 0.5, 0.01)
s21 = 1 - 2 * y01 ** 2 - 1/2 * (1 - y01) ** 2

y02 = np.arange (0.5, 1.01, 0.01)
s22 = 3/2 * (1-y02)** 2

s2 = np.concatenate ((s21, s22),axis=0)

plt.plot(y0, s1, y0, s2)
plt.legend(['$s_1$','$s_2$'])
plt.xlabel('$y_0$')
plt.ylabel('$s$')
plt.show()
```

So the 1st form has a smaller feasible space. This tightness costs 1 more constraint.
